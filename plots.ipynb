```python
import matplotlib.pyplot as plt 
```

```python
#list of coordinates x and y
# you can use and define values and passed them instead of numbers
x_coords=[0,1,2,3,4]
y_coords=[1,3,2,5,3]

#build the line graph

plt.plot(x_coords, y_coords)

#display the line graph
plt.show()
```

```python
#limit the x and y coordinates
plt.plot(x_coords, y_coords)
plt.xlim(xmin=0, xmax=10)
plt.ylim(ymin=0,ymax=6)
plt.show()
```

```python
#customize the ticks or range of values, by 2, by 1, by 4
#to change ticks, put the regular value and the next in '' 
plt.plot(x_coords, y_coords)  #plot is a function 
years=['2016','2017','2018','2019','2020']
plt.xticks([0,1,2,3,4],years) #changes numbers to years

plt.yticks([1,2,3,4,5],['$0m','$1m','$2m','$3m','$4m','$5m'])

plt.show() #function to show the plot
```

```python
#Add Labels and Grid
#add points

plt.plot(x_coords, y_coords, marker='o') #marker is a dot for points
plt.xticks([0,1,2,3,4],['2016','2017','2018','2019','2020']) #changes numbers to years

plt.yticks([1,2,3,4,5],['$0m','$1m','$2m','$3m','$4m','$5m'])

#Add a title.
plt.title('Sales by Year')  #use title function

#Add labels to axes, uses label function 
plt.xlabel('Year')
plt.ylabel('Sales')

#add grid
plt.grid(True)  #use grid function and algorithm true 
plt.show()

```

```python
#Bar Charts
plt.bar(x_coords, y_coords)   #instead of plot, call bar for chart
plt.show()
```

```python
#Change Bar Width
bar_width=0.5 #you can set a variable to change it
plt.bar(x_coords, y_coords, bar_width) #width is now 1
plt.show()
```

```python
#colors for variables 
'b'=Blue
'g'=Green
'r'=Red
'c'=Cyan
'm'=Magenta
'y'=Yellow
'k'=Black
'w'=White

```

```python
plt.bar(x_coords,y_coords, color=('r','g','b','c','k')) 
#bars by 1, 2, to last 
plt.show()
```

```python
#add the same of plots to bars
plt.bar(x_coords,y_coords, color=('r','g','b','c','k')) 

plt.xticks([0,1,2,3,4],['2016','2017','2018','2019','2020']) #changes numbers to years

plt.yticks([1,2,3,4,5],['$0m','$1m','$2m','$3m','$4m','$5m'])

#Add a title.
plt.title('Sales by Year')  #use title function

#Add labels to axes, uses label function 
plt.xlabel('Year')
plt.ylabel('Sales')
```

```python
#Pie Chart, uses .pie function instead of plot or bar

plt.pie(y_coords)
plt.show()
```

```python
#Add labels and years

years=[2016,2017,2018,2019,2020]
plt.pie(y_coords, labels=years) #make sure to put labels as = years
plt.title('Sales by Year')
plt.show()
```
